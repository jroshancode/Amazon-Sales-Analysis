# Min-Max and Z_score normalization 

```python
Q1 = data['Pulse'].quantile(0.25)
Q3 = data['Pulse'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = data[(data['Pulse'] < lower_bound) | (data['Pulse'] > upper_bound)]
print("Outliers (IQR method):", outliers_iqr)
```

```python
df = pd.DataFrame({'data': [1, 2, 3, 4, 5, 100, 6, 7, 8]})
df.boxplot(column='data')
plt.show()
```

```python
Q1 = df['data'].quantile(0.25)
Q3 = df['data'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = df[(df['data'] < lower_bound) | (df['data'] > upper_bound)]
print("Outliers (IQR method):", outliers_iqr)
```

```python
df_cleaned = df[(df['data'] >= lower_bound) & (df['data'] <= upper_bound)]
print("Data after removing outliers:\n", df_cleaned)
```

```python
from scipy.stats import zscore
df['z_score'] = zscore(df['data'])
outliers_zscore = df[abs(df['z_score']) > 2] # Threshold can be adjusted
print("Outliers (Z-score method):", outliers_zscore)
```

```python
from datetime import datetime, timedelta

# Create a timedelta object for a duration of 5 days, 3 hours, and 30 minutes
duration = timedelta(days=5, hours=3, minutes=30)
print(f"Duration: {duration}")

# Get the current date and time
now = datetime.now()
print(f"Current time: {now}")

# Calculate a future date by adding a timedelta
future_date = now + timedelta(days=10)
print(f"Date 10 days from now: {future_date}")

# Calculate a past date by subtracting a timedelta
past_date = now - timedelta(weeks=2)
print(f"Date 2 weeks ago: {past_date}")

# Calculate the difference between two datetime objects
start_time = datetime(2025, 1, 1, 10, 0, 0)
end_time = datetime(2025, 1, 5, 14, 30, 0)
time_elapsed = end_time - start_time
print(f"Time elapsed: {time_elapsed}")
print(f"Total seconds elapsed: {time_elapsed.total_seconds()}")
```
